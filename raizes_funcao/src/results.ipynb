```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Avaliação dos resultados

```python
df = pd.read_csv('../output/methods_result_2.csv', sep=',')
df['method'] = df['method'].replace({
    'Bissection': 'Bissecção',
    'False Position': 'Falsa Posição',
    'Fixed Point': 'Ponto Fixo',
    'Newton Raphson': 'Newton-Raphson',
    'Secant': 'Secante'
})
df.head()
```

### 1ª raiz

```python
df_1 = df[(df['function'] == '(x/2) - tan(2*x)') & (df['root_index'] == 0) & (df['iter'] <= 100)]
df_1 = df_1[(df_1['method'] != 'Ponto Fixo')]

fig = plt.figure(figsize=(10, 6))
sns.lineplot(data=df_1, x='iter', y='relative_error', hue='method', palette='Set1')
#plt.title('Convergência dos métodos para f(x) = (x/2) - tan(2*x)')
plt.xlabel('Iteração', fontsize=14)
plt.xticks(fontsize=12)
plt.ylabel('Erro estimado relativo (%)', fontsize=14)
plt.legend(title='Método', loc='upper right', fontsize=11, title_fontsize=12)
plt.show()
```
