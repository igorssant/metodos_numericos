```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
powers = []
euler_hats = []
errors = []

with open('../output/euler_results.txt', 'r') as fd:
    for line in fd.readlines()[2:]:
        tmp = line.split()

        powers.append(np.int32(tmp[0].strip()))
        euler_hats.append(np.float64(tmp[1].strip()))
        errors.append(np.float64(tmp[2].strip()))

df = pd.DataFrame({
    'Potencia': powers,
    'Euler calculado': euler_hats,
    'Erro (%)': errors
})

df.to_parquet("../output/error_aproximacao_euler", index=False)
df.to_csv("../output/error_aproximacao_euler.csv", index=False, float_format='%.13f')

df.head()
```

```python
fig = plt.figure()

plt.title("Erro da aproximação de Euler com potência de dez")
plt.plot(powers, errors)
plt.xlabel("Potência")
plt.ylabel("Erro relativo")
plt.savefig('../output/error_aproximacao_euler.png')
plt.show()
```
